```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tic
```

```python
kpoints = np.genfromtxt('tmp_kpoint.txt', usecols=(0,1,2))
#print(kpoints)
```

```python
# Hexagonal lattice
a1 = np.array([np.sqrt(3.0), 0.0, 0.0])
a2 = np.array([0.5*np.sqrt(3.0), 1.5, 0.0])
a3 = np.array([0,0,1])
v = np.dot(a1,np.cross(a2,a3))
b1 = 2.0*np.pi*np.cross(a2,a3)/v
b2 = 2.0*np.pi*np.cross(a3,a1)/v
M = 0.5*(b2)
Kprime = (2*b2+b1)/3
K = (2*b1+b2)/3
#G_list = np.array([[0,0,0],b1,b2,-b1,-b2,b1+b2,-(b1+b2),(b1-b2),-(b1-b2)])
G_list = np.array([[0,0,0],M,Kprime,K])
```

```python
#print(b2)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.set_aspect('equal', adjustable='box')

ax.plot(kpoints[:,1],kpoints[:,2],label='',ls='',marker='o',c='r',lw=3, ms=4)

for i in range(len(G_list)):
    G = G_list[i]
    ax.plot(G[0],G[1],label='',ls='',marker='o',c='b',lw=3, ms=10)
    

```
